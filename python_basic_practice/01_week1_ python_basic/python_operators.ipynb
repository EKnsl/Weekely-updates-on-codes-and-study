# Arithmetic operators in Python

```python
x = 15
y = 4

print("x = ", x)
print("y = ", y)

print("\n")
# Output: x + y = 19
print('x + y =',x+y)

# Output: x - y = 11
print('x - y =',x-y)

# Output: x * y = 60
print('x * y =',x*y)

# Output: x / y = 3.75
print('x / y =',x/y)

# Output: x // y = 3 , // performs integer division
print('x // y =',x//y)

# Output: x ** y = 50625, perofrms x to the power y
print('x ** y =',x**y)
```

# Comparison operators in Python

```python
x = 10
y = 12

print("x = ", x)
print("y = ", y)

print("\n")

# Output: x > y is False
print('x > y is',x > y)

# Output: x < y is True
print('x < y is',x < y)

# Output: x == y is False
print('x == y is',x == y)

# Output: x != y is True
print('x != y is',x != y)

# Output: x >= y is False
print('x >= y is',x >= y)

# Output: x <= y is True
print('x <= y is',x <= y)
```

# Logical Operators in Python

```python
x = True
y = False

print("x = ", x)
print("y = ", y)
print("\n")

print('x and y is',x and y)

print('x or y is',x or y)

print('not x is',not x)
```

# Identity operators in Python

```python
x1 = 5
y1 = 5
x2 = 'Hello'
y2 = 'Hello'
x3 = [1,2,3]
y3 = [1,2,3]

print("x1 = ", x1)
print("y1 = ", y1)
print("x2 = ", x2)
print("y2 = ", y2)
print("x3 = ", x3)
print("y3 = ", y3)
print("\n")

# Output: False
print(x1 is not y1)

# Output: True
print(x2 is y2)

# Output: False
print(x3 is y3)
```
